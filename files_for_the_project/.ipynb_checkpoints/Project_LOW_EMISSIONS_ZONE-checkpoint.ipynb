#### 4. Split the data from air pollutant in new columns corresponding to NO₂, O₃, PM2,5 and PM10. 

```python
# Initialize new columns with NaN values
dfb['NO2_pollution($\\mu$g/m$^3$)'] = np.nan
dfb['O3_pollution($\\mu$g/m$^3$)'] = np.nan
dfb['PM10_pollution($\\mu$g/m$^3$)'] = np.nan
dfb['PM2.5_pollution($\\mu$g/m$^3$)'] = np.nan

# Fill the new columns with corresponding pollution values
for index, row in dfb.iterrows():
    pollutant = row['air_pollutant']
    pollution_value = row['air_pollution_population_weighted_average_($\mu$g/m$^3$)']
    dfb.at[index, f'{pollutant}_pollution($\\mu$g/m$^3$)'] = pollution_value
    
```

```python
dfb.drop(['air_pollution_population_weighted_average_($\mu$g/m$^3$)', 'air_pollutant'], axis=1, inplace=True)
```

```python
# Initialize new columns with NaN values
dfw['NO2_pollution($\\mu$g/m$^3$)'] = np.nan
dfw['O3_pollution($\\mu$g/m$^3$)'] = np.nan
dfw['PM10_pollution($\\mu$g/m$^3$)'] = np.nan
dfw['PM2.5_pollution($\\mu$g/m$^3$)'] = np.nan

# Fill the new columns with corresponding pollution values
for index, row in dfw.iterrows():
    pollutant = row['air_pollutant']
    pollution_value = row['air_pollution_population_weighted_average_($\mu$g/m$^3$)']
    dfw.at[index, f'{pollutant}_pollution($\\mu$g/m$^3$)'] = pollution_value
```

#### 5. Finding and dropping outliers in O₃_pollution 

```python
def outliers(column, threshold = 3):
   
    data = column[abs(column.apply(lambda x: (x - column.mean()) / column.var() ** (1/2))) > threshold]
    
    return data

O3_outliers_b = outliers(dfb["O3_pollution($\mu$g/m$^3$)"]) # From Erin deleting outliers 
len(O3_outliers_b)
```

```python
O3_outliers_b
```

```python
def remove_outliers(data, outliers):
    return data.drop(outliers.index)

# Remove outliers from the DataFrame
dfb_no_outliers = remove_outliers(dfb, O3_outliers_b)

# Verify the removal
dfb_no_outliers.shape
```

```python
def outliers(column, threshold = 2):
   
    data = column[abs(column.apply(lambda x: (x - column.mean()) / column.var() ** (1/2))) > threshold]
    
    return data

O3_outliers_w = outliers(dfw["O3_pollution($\mu$g/m$^3$)"]) # From Erin deleting outliers 
len(O3_outliers_w)
```

## LINEAR REGRESSION MODEL USING BERLIN DATAFRAME

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler 
from sklearn.preprocessing import StandardScaler
```

```python
transformer = MinMaxScaler().fit(dfb_numerical_cleaned)
x_normalized = transformer.transform(dfb_numerical_cleaned)
```

```python
x_norm = pd.DataFrame(x_normalized, columns = dfb_numerical_cleaned.columns)
x_norm.head()
```

```python
# Eliminar filas con valores faltantes en la variable objetivo y
dfb_numerical_cleaned.dropna(subset=['NO2_pollution($\\mu$g/m$^3$)'], inplace=True)

# Definir la variable objetivo y las características
y = dfb_numerical_cleaned['NO2_pollution($\\mu$g/m$^3$)']
X = dfb_numerical_cleaned.drop('NO2_pollution($\\mu$g/m$^3$)', axis=1)  # Eliminar la variable objetivo

# Escalar las características
scaler = MinMaxScaler()
x_normalized = scaler.fit_transform(X)
x_norm = pd.DataFrame(x_normalized, columns=X.columns)

# Dividir los datos en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(x_norm, y, test_size=0.2, random_state=2222)
```

```python
# Crear e inicializar el modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, y_train)
```

```python
# Evaluar el modelo
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

print("Coeficiente de determinación (R^2) en conjunto de entrenamiento:", train_score)
print("Coeficiente de determinación (R^2) en conjunto de prueba:", test_score)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

## HYPOTESIS TEST FOR BERLIN DATABASE

Using our data for Berlin, for the test we're using the cleaned data not normalized. 

```python
import scipy.stats

NO2_before_2010 = dfb_numerical_cleaned[dfb_numerical_cleaned['year'] < 2010]['NO2_pollution($\\mu$g/m$^3$)']
NO2_from_2013 = dfb_numerical_cleaned[dfb_numerical_cleaned['year'] >= 2013]['NO2_pollution($\\mu$g/m$^3$)']

confidence_level = 0.95

degrees_freedom = len(NO2_before_2010) - 1 


sample_mean = np.mean(NO2_before_2010)

sample_standard_error = scipy.stats.sem(NO2_before_2010)

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)


print(confidence_interval)
```
