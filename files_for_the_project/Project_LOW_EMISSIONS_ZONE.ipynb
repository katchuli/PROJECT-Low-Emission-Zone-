#### 5. Split the data from air pollutant in new columns corresponding to NO₂, O₃, PM2,5 and PM10. 

```python
# Initialize new columns with NaN values
dfb['NO2_pollution($\\mu$g/m$^3$)'] = np.nan
dfb['O3_pollution($\\mu$g/m$^3$)'] = np.nan
dfb['PM10_pollution($\\mu$g/m$^3$)'] = np.nan
dfb['PM2.5_pollution($\\mu$g/m$^3$)'] = np.nan

# Fill the new columns with corresponding pollution values
for index, row in dfb.iterrows():
    pollutant = row['air_pollutant']
    pollution_value = row['air_pollution_population_weighted_average_($\mu$g/m$^3$)']
    dfb.at[index, f'{pollutant}_pollution($\\mu$g/m$^3$)'] = pollution_value
    
```

#### 6. Finding and dropping outliers in O₃_pollution 

```python
def outliers(column, threshold = 3):
   
    data = column[abs(column.apply(lambda x: (x - column.mean()) / column.var() ** (1/2))) > threshold]
    
    return data

O3_outliers_b = outliers(dfb["O3_pollution($\mu$g/m$^3$)"]) # From Erin deleting outliers 
len(O3_outliers_b)
```

```python
O3_outliers_b
```

```python
def remove_outliers(data, outliers):
    return data.drop(outliers.index)

# Remove outliers from the DataFrame
dfb_cleaned = remove_outliers(dfb, O3_outliers_b)

# Verify the removal
dfb_cleaned.shape
```

```python
def outliers(column, threshold = 2):
   
    data = column[abs(column.apply(lambda x: (x - column.mean()) / column.var() ** (1/2))) > threshold]
    
    return data

O3_outliers_w = outliers(dfw["O3_pollution($\mu$g/m$^3$)"]) # From Erin deleting outliers 
len(O3_outliers_w)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler 
from sklearn.preprocessing import StandardScaler
```

```python
#Scaling 
transformer = MinMaxScaler().fit(numeric_b_imputed)
x_normalized = transformer.transform(numeric_b_imputed)
```

```python
x_norm = pd.DataFrame(x_normalized, columns = numeric_b_imputed.columns)
x_norm.head()
```

```python

# Eliminar filas con valores faltantes en la variable objetivo y
numeric_b_imputed.dropna(subset=['NO2_pollution($\\mu$g/m$^3$)'], inplace=True)

# Definir la variable objetivo y las características
y = numeric_b_imputed['NO2_pollution($\\mu$g/m$^3$)']
X = numeric_b_imputed.drop('NO2_pollution($\\mu$g/m$^3$)', axis=1)  # Eliminar la variable objetivo

# Escalar las características
scaler = MinMaxScaler()
x_normalized = scaler.fit_transform(X)
x_norm = pd.DataFrame(x_normalized, columns=X.columns)

# Dividir los datos en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(x_norm, y, test_size=0.2, random_state=2222)

# Crear e inicializar el modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, y_train)

# Evaluar el modelo
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

print("Coeficiente de determinación (R^2) en conjunto de entrenamiento:", train_score)
print("Coeficiente de determinación (R^2) en conjunto de prueba:", test_score)
```
