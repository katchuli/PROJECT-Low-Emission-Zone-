## EXAMINE THE DATA

#### 1. Exporting the file data from two cities, Berlin(with LEZ since 2008) and Warsaw (will implemeted in 2024). 

```python
import pandas as pd 
import numpy as np
```

```python
# Reading data into a DataFrame
df1 = pd.read_csv('Berlin_emissions.csv')
```

```python
df2 = pd.read_csv('Warsaw_emissions.csv')
```

#### 2. Combining the two DataFrames using merge()

```python
df3 = pd.merge(df1, df2, how='outer')
```

```python
display(df3)
```

## DATA CLEANING

#### 1. Delete star, lower case with underscore

```python
df3 = df3.rename(columns={'Total City Population *': 'Total City Population'})
```

```python
cols = []
for col in df3.columns:
    cols.append(col.lower())
df3.columns = cols
```

```python
df3 = df3.rename(columns={'populated area [km2]': 'populated area (Km$^2$)', 'air pollution average [ug/m3]': 'air pollution average ($\mu g$/m$^3$)', 'air pollution population weighted average [ug/m3]': 'air pollution population weighted average ($\mu g$/m$^3$)', 'premature deaths - lower ci': 'premature deaths - lower CI', 'premature deaths - upper ci': 'premature deaths - upper CI', 'years of life lost - lower ci': 'years of life lost - lower CI', 'years of life lost - upper ci': 'years of life lost - upper CI'
  })
```

```python
df3.columns = df3.columns.str.replace(' ', '_')
```

```python
df3
```

#### 2. Check NaN values

```python
df3.isna().sum()
```

In principle this values are not relevant to our study. Checking in detalil both dat sets separated there are same NaN values for Berlin and Warsaw but if we delete it will lose data related with pollutant particles PM10 so we keep it. 

#### 3. Dropping unnecessary columns 

As categorical data we're only interested in the country and city, not the city code or the boundary specification, either the health risk scenario. For that, this columns are deleted and a new dataframe created called d4.

```python
df3 = df3.drop(['city_boundary_specification_(lau/grid)', 'city_code', 'health_risk_scenario'], axis=1)
```

#### 4. Check duplicates

```python
duplicates = df3.duplicated()
print(duplicates)
```

 There are no duplicates

```python
import numpy as np

# Creamos columnas separadas para cada clase de contaminante y las inicializamos con NaN
df3['NO2_pollution'] = np.nan
df3['O3_pollution'] = np.nan
df3['PM10_pollution'] = np.nan
df3['PM2.5_pollution'] = np.nan

# Asignamos los valores de 'air_pollution_average' a las columnas correspondientes según el tipo de contaminante
df3.loc[df3['air_pollutant'].str.contains('NO2', case=False), 'NO2'] = df3['air_pollution_average_($\\mu_g$/m$^3$)']
df3.loc[df3['air_pollutant'].str.contains('O3', case=False), 'O3'] = df3['air_pollution_average_($\\mu_g$/m$^3$)']
df3.loc[df3['air_pollutant'].str.contains('PM10', case=False), 'PM10'] = df3['air_pollution_average_($\\mu_g$/m$^3$)']
df3.loc[df3['air_pollutant'].str.contains('PM2.5', case=False), 'PM2.5'] = df3['air_pollution_average_($\\mu_g$/m$^3$)']

# Mostramos el DataFrame resultante
display(df3)
```
