## DATA CLEANING

#### 1. Delete star, lower case with underscore

```python
dfb = dfb.rename(columns={'Total City Population *': 'Total City Population'})
```

```python
dfw = dfw.rename(columns={'Total City Population *': 'Total City Population'})
```

```python
cols = []
for col in dfb.columns:
    cols.append(col.lower())
dfb.columns = cols
```

```python
cols = []
for col in dfw.columns:
    cols.append(col.lower())
dfw.columns = cols
```

```python
dfb = dfb.rename(columns={'populated area [km2]': 'populated area (Km$^2$)', 'air pollution average [ug/m3]': 'air pollution average ($\mu g$/m$^3$)', 'air pollution population weighted average [ug/m3]': 'air pollution population weighted average ($\mu g$/m$^3$)', 'premature deaths - lower ci': 'premature deaths - lower CI', 'premature deaths - upper ci': 'premature deaths - upper CI', 'years of life lost - lower ci': 'years of life lost - lower CI', 'years of life lost - upper ci': 'years of life lost - upper CI'
  })
```

```python
dfw = dfw.rename(columns={'populated area [km2]': 'populated area (Km$^2$)', 'air pollution average [ug/m3]': 'air pollution average ($\mu g$/m$^3$)', 'air pollution population weighted average [ug/m3]': 'air pollution population weighted average ($\mu g$/m$^3$)', 'premature deaths - lower ci': 'premature deaths - lower CI', 'premature deaths - upper ci': 'premature deaths - upper CI', 'years of life lost - lower ci': 'years of life lost - lower CI', 'years of life lost - upper ci': 'years of life lost - upper CI'
  })
```

```python
dfb.columns = dfb.columns.str.replace(' ', '_')
```

```python
dfw.columns = dfw.columns.str.replace(' ', '_')
```

#### 3. Dropping unnecessary columns 

As categorical data we're only interested in the country and city, not the city code or the boundary specification, either the health risk scenario. For that, this columns are deleted and a new dataframe created called d4.

```python
dfb = dfb.drop(['city_boundary_specification_(lau/grid)', 'city_code', 'health_risk_scenario'], axis=1)
```

```python
dfw = dfw.drop(['city_boundary_specification_(lau/grid)', 'city_code', 'health_risk_scenario'], axis=1)
```

#### 4. Check duplicates

```python
duplicates = dfb.duplicated()
print(duplicates)
```

```python
dfb = dfb.drop_duplicates()
```

```python
duplicates = dfw.duplicated()
print(duplicates)
```

```python
dfw = dfw.drop_duplicates()
```

#### 5. Split the data from air pollutant in new columns corresponding to NO2, O3, PM2,5 and PM10. 

```python
# Initialize new columns with NaN values
dfb['NO2_pollution($\\mu_g$/m$^3$)'] = np.nan
dfb['O3_pollution($\\mu_g$/m$^3$)'] = np.nan
dfb['PM10_pollution($\\mu_g$/m$^3$)'] = np.nan
dfb['PM2.5_pollution($\\mu_g$/m$^3$)'] = np.nan

# Fill the new columns with corresponding pollution values
for index, row in dfb.iterrows():
    pollutant = row['air_pollutant']
    pollution_value = row['air_pollution_average_($\\mu_g$/m$^3$)']
    dfb.at[index, f'{pollutant}_pollution($\\mu_g$/m$^3$)'] = pollution_value
```

```python
display(dfb)
```

```python
# Initialize new columns with NaN values
dfw['NO2_pollution($\\mu_g$/m$^3$)'] = np.nan
dfw['O3_pollution($\\mu_g$/m$^3$)'] = np.nan
dfw['PM10_pollution($\\mu_g$/m$^3$)'] = np.nan
dfw['PM2.5_pollution($\\mu_g$/m$^3$)'] = np.nan

# Fill the new columns with corresponding pollution values
for index, row in dfw.iterrows():
    pollutant = row['air_pollutant']
    pollution_value = row['air_pollution_average_($\\mu_g$/m$^3$)']
    dfw.at[index, f'{pollutant}_pollution($\\mu_g$/m$^3$)'] = pollution_value
```

```python
display(dfw)
```
